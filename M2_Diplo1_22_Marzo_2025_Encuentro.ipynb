## El problema del encuentro

Romeo y Julieta pretenden encontrarse en un lugar específico entre $[0,60]$.

Si Romeo llega primero, entonces piensa esperar 10 minutos, tmb Julieta

Se elige un número al azar entre 0 y 60, y supongamos que el par de números se elige sobre el cuadrado $60^2$, de modo que la proba de evento, sea el área favorable entre el área total

Sea el evento:
a) Romeo y Julieta se encuentran a una distancia de $10$ minutos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Se generan 100,000 simulaciones
muestra = 10000

# Generación de tiempos aleatorios
# x -> tiempo de Romeo [0,60] (notación de la miss)
# y -> tiempo de Julieta [0,60]

t_romeo = np.random.uniform(0,60,muestra)
t_julieta = np.random.uniform(0,60,muestra)

# Condición de encuentro
## red = r = se encuentran
## blue = b = no se encuentran

encuentro = np.where(abs(t_julieta-t_romeo)<=10,'r','b')

# Proba del encuentro
k=0
for i in range(muestra):
    if encuentro[i]=='r':
        k+=1
print("La proba del evento A es:", k/muestra)

# Gráfica
plt.scatter(t_julieta,t_romeo,c=encuentro,s=1)
plt.show()
```
